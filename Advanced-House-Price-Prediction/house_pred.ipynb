```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns 

#display all the columns of the dataframe

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv('train.csv')
print(dataset.shape)
```

```python
dataset.head() 
```

In data analysis what we have to do : <br>
1.Missing values <br>
2.All the numerical values<br>
3.Distribution of the numerical variables <br>
4.Categorical variables

```python
features_with_na = [cheese for cheese in dataset.columns if dataset[cheese].isnull().sum()>1]

for cheese in features_with_na :
    print(cheese, np.round(dataset[cheese].isnull().mean(),4), '%missing values') 
```

```python
for cheese in features_with_na :
    data = dataset.copy()
    data[cheese] = np.where(data[cheese].isnull(),1,0)

    data.groupby(cheese)['SalePrice'].median().plot.bar()
    plt.title(cheese)
    plt.show()
```
